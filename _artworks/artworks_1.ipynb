```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

# 1. 매개변수 설정
n = 10  
d = 9
k = n / d

# 2. 각도(theta) 배열 생성
# 0부터 시작해서 패턴이 완전히 닫힐 때까지 충분한 범위를 설정합니다. (d * 2 * pi)
theta = np.linspace(0, d * 2 * np.pi, 1000)

# 3. 로즈 커브의 극좌표 방정식 적용
r = np.cos(k * theta)

# 4. 극좌표를 데카르트 좌표(x, y)로 변환
x = r * np.cos(theta)
y = r * np.sin(theta)

# 5. 예술적인 시각화
plt.figure(figsize=(8, 8), facecolor='white') # 캔버스 크기와 배경색 설정
ax = plt.subplot(111, facecolor='white')

# 부드러운 네온 느낌의 색상으로 선 그리기
ax.plot(x, y, color='red', linewidth=1.5, alpha=0.8)

# 수학적 그래프의 느낌을 없애기 위해 축 숨기기
ax.axis('equal') # x축과 y축의 비율을 동일하게 맞춰 찌그러지지 않게 함
ax.axis('off')

plt.title(f"Mathematical Rose (n={n}, d={d})", color='white', fontsize=14, pad=20)
plt.tight_layout()
plt.show()
```
